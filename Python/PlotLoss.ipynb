```python
import numpy as np
from scipy import io
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
from scipy import io
import mpld3
mpld3.enable_notebook()
```

## Experiment 1 
Description of experiment

```python
expt1=io.loadmat('expt1/output.mat')
loss=expt1['loss']
loss = loss[loss>0]
lossall = loss[::100]
tall = np.arange(len(lossall))*100

print 'Last 5 full loss iterations'
print lossall[-5:]

plt.plot(loss,'.');
plt.plot(tall,lossall,'ko-');
plt.legend(('Subsampled loss','Full loss'));
plt.ylim(0,3e6);
plt.xlabel('iteration');
plt.ylabel('MSE');
```

```python
expt1=io.loadmat('expt1/output.mat')
loss_gt = expt1['loss_gt']
loss_gt = loss_gt[loss_gt>0]
loss_gtall = loss_gt[::100]
tall = np.arange(len(loss_gtall))*100

print 'Last 5 full loss iterations vs ground truth'
print loss_gtall[-5:]

plt.plot(loss_gt,'.');
plt.plot(tall,loss_gtall,'ko-');
plt.legend(('Subsampled loss','Full loss GT'));
plt.ylim(0,2*np.amax(loss_gtall));
plt.xlabel('iteration');
plt.ylabel('MSE');
```
