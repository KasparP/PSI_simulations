```python
import numpy as np
from scipy import io
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
from scipy import io
```

```python
expt1=io.loadmat('expt1/output.mat')
loss=expt1['loss']
loss = loss[loss>0]
lossall = loss[::100]
tall = np.arange(len(lossall))*100

print 'Last 5 full loss iterations'
print lossall[-5:]

plt.plot(loss,'.');
plt.plot(tall,lossall,'ko-');
plt.legend(('Subsampled loss','Full loss'));
plt.ylim(0,3e6);
plt.xlabel('iteration');
plt.ylabel('MSE');
```
